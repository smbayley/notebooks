```python
data = pd.read_csv('./data/travel-vs-laser-vs-sphere.csv').as_matrix()
error = np.abs(data[:, 0].reshape(-1, 1) - data[:, 1:])
```

```python
df = pd.DataFrame(error[:, 0], index=data[:, 0], columns=('error',))
ax = df.plot(legend=False)
ax.set_xlabel('distance (m)')
ax.set_ylabel('error (m)')
plt.title('Ellipsoidal Laser Distance vs Ellipsoidal Travel Distance')
plt.show()
```

```python
df = pd.DataFrame(error, index=data[:, 0], columns=('ellipsoid laser', 'surface distance, spherical approx'))
ax = df.plot()
ax.set_xlabel('distance (m)')
ax.set_ylabel('error (m)')
plt.title('Ellipsoidal Laser Distance vs Spherical Surface Distance')
plt.show()
```
